## Exercise 2 :  Tidying a collaboration

```python
month_no = {'jan': '01', 'feb': '02', 'mar': '03',
            'apr': '04', 'may': '05', 'jun': '06',
            'jul': '07', 'aug': '08', 'sep': '09',
            'oct': '10', 'nov': '11', 'dec': '12'}

def conversion(file_name):
    month = file_name.split("-")[0]
    remaining = file_name.split("-")[1]
    left = remaining.split("_")[0]
    right = remaining.split("_")[1]
    return f"{month_no[month]}-{left}-{right}"

conversion('oct-08_singapore.txt')

# My first thought was already a function, will use the same for next part. 
```

```python
def rename_my_file(old_file_name):
    month = old_file_name.split("-")[0]
    remaining = old_file_name.split("-")[1]
    left = remaining.split("_")[0]
    right = remaining.split("_")[1]
    return f"{month_no[month]}-{left}-{right}"

conversion('oct-08_singapore.txt')
```

```python
old_file_name = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']

new_file_name = []
for file in old_file_name:
    new_file = rename_my_file(file)
    new_file_name.append(new_file)
print(new_file_name)

# Note: append can only mutate but cannot create new list: 
# I renamed the variables from new-file-name to new_file_name as - is not acceptable variable name.
```
